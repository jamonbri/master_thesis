### Simulation Run

```python
# Imports

import pandas as pd
from model.model import RecommenderSystemModel
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
# Initialize model

rec_sys = RecommenderSystemModel(n_users=100, dummy=False, steps=10, priority=None, seed=123)
```

### Data Analysis

```python
# Get vector differences

results_filepath = [filepath for filepath in rec_sys.csv_filepaths if filepath.split("/")[-1].startswith("run")]
df_vector_diff = rec_sys.results.get_vector_diff_df(results_filepath[0])
```

```python
# Plot vector differences

df_vector_diff["vector_diff"].hist(
    bins=20, 
    grid=False, 
    edgecolor="black"
)
plt.xlabel("Cosine similarity")
plt.ylabel("Frequency")
plt.title("Histogram of cosine similarity between first and last vector per user")
```
