### Simulation Run

```python
# Imports

import pandas as pd
from model.model import RecommenderSystemModel
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
# Initialize benchmark model

rec_sys = RecommenderSystemModel(n_users=100, dummy=False, steps=200, priority=None, seed=12345)
```

```python
# Run benchmark model

rec_sys.run_model()
```

```python
# Initialize random model at 10%

rec_sys_random_10 = RecommenderSystemModel(n_users=100, dummy=False, steps=200, priority=0.1, seed=12345)
```

```python
# Run random model at 10%

rec_sys_random_10.run_model()
```

```python
# Initialize 'thriller' model

rec_sys_fantasy = RecommenderSystemModel(n_users=100, dummy=False, steps=200, priority="fantasy", seed=12345)
```

```python
# Run 'fantasy' model

rec_sys_fantasy.run_model()
```

### Data Analysis

```python
# Get vector differences for benchmark model

results_filepath = [filepath for filepath in rec_sys.csv_filepaths if filepath.split("/")[-1].startswith("run")]
df_vector_diff = rec_sys.results.get_vector_diff_df(results_filepath[0])
```

```python
# Plot vector differences for benchmark model

df_vector_diff["vector_diff"].hist(
    bins=20, 
    grid=False, 
    edgecolor="black"
)
plt.xlabel("Cosine similarity")
plt.ylabel("Frequency")
plt.title("Histogram of cosine similarity between first and last vector per user (benchmark)")
```
